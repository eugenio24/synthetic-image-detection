```python
import os
import glob
import time
from PIL  import Image
from datetime import datetime
import random

date = datetime.now().strftime("%Y-%m-%d_%H%M%S")
```

```python
stylegan3_metfaces_config = ["stylegan3-r-metfacesu-1024x1024", "stylegan3-t-metfacesu-1024x1024"]
config = stylegan3_metfaces_config[1]

fake_folder = f"/content/drive/MyDrive/datasets/my_generated/stylegan3/{config}"
real_folder = "/content/drive/MyDrive/datasets/metfaces-release/images"

weights_path = './weights/gandetection_resnet50nodown_stylegan2.pth'

out_folder = "/content/drive/MyDrive/signal_prj/stylegan3/metfaces"
```

```python
from resnet50nodown import resnet50nodown

from torch.cuda import is_available as is_available_cuda
device = 'cuda:0' if is_available_cuda() else 'cpu'

net = resnet50nodown(device, weights_path)
```

```python
fake = sum([glob.glob(os.path.join(fake_folder, '*.png'))], list())[-2000:]
real = sum([glob.glob(os.path.join(real_folder, '*.png'))], list())[-336:] 

imgs = fake + real
random.shuffle(imgs)
imgs_len = len(imgs)
```

```python
def run():
  output_csv = f'{out_folder}/out.{date}.{config}.csv'    
  
  with open(output_csv,'w') as fid:
      fid.write('filename,time,fake,logit\n')
      fid.flush()

      for index, filename in enumerate(imgs):          
          isFake = "stylegan3" in filename

          tstart = time.time()
          img = Image.open(filename).convert('RGB')
          img.load()
          logit = net.apply(img)
          tend = time.time()
        
          outstr = '%s,%f,%i,%f\n' % (filename, tend-tstart, isFake, logit)
          print(outstr)
          fid.write(outstr)
          fid.flush()
```

```python
run()
```
